```
normalized_eval_path = '/content/drive/MyDrive/Colab Notebooks/regaetton_songs_nlp/data/normalized_eval_lyrics.csv'
normalized_train_path = '/content/drive/MyDrive/Colab Notebooks/regaetton_songs_nlp/data/normalized_train_lyrics.csv'
```

```
import pandas as pd

from gensim.models import Word2Vec
```

# Get the data

```
train = pd.read_csv(normalized_train_path)
eval = pd.read_csv(normalized_eval_path)
```

# Build our training data

We are going to need a list of lists of tokens. The upper list is the whole training data, and the lists inside are the lyrics for each song splitted into tokens.

```
corpus = list()

def create_list_tokens(lyrics):
    global corpus

    list_tokens = lyrics.split()
    corpus.append(list_tokens)


train.lyrics.apply(create_list_tokens)
eval.lyrics.apply(create_list_tokens)
    

```

```
print(f'Amount of lyrics: {len(corpus)}')
```

# Training

We are going to train both architectures: CBOW and skip-gram

```
cbow_model = Word2Vec(corpus, size=300, min_count=1, sg=0)
skip_model = Word2Vec(corpus, size=300, min_count=1, sg=1)
```

```
print("Similarity between cama and comerte:",cbow_model.wv.similarity('cama', 'comerte'))
print("Similarity between cama and auto:",cbow_model.wv.similarity('cama', 'auto'))
```

```
print("Similarity between cama and comerte:",skip_model.wv.similarity('cama', 'comerte'))
print("Similarity between cama and auto:",skip_model.wv.similarity('cama', 'auto'))
```

```
cbow_model.wv.most_similar('cama')
```

```
skip_model.wv.most_similar('cama')
```

Apparently, the embeddings that describes our task (detect sexual content) is the cbow model.
