```
cbow_path = '/content/drive/MyDrive/Colab Notebooks/regaetton_songs_nlp/embeddings/cbow_model.bin'
```

```
import io
from gensim.models import Word2Vec
```

```
cbow_model = Word2Vec.load(cbow_path)
```

```
words_file = 'words'
embeddings_file = 'lyrics_embeddings'

out_words = io.open(f'{words_file}.tsv', 'w', encoding='utf-8')
out_vectors = io.open(f'{embeddings_file}.tsv', 'w', encoding='utf-8')

```

```
# Write meta file and vector file
for index in range(len(cbow_model.wv.index2word)):
    word = cbow_model.wv.index2word[index]
    vec = cbow_model.wv.vectors[index]
    out_words.write(word + "\n")
    out_vectors.write('\t'.join([str(x) for x in vec]) + "\n")
out_words.close()
out_vectors.close()
```
